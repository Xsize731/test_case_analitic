```python
# переведем столбцы в числовой формат 
column_names = ['Ноябрь 2022', 'Декабрь 2022', 'Январь 2023',
       'Февраль 2023', 'Март 2023', 'Апрель 2023', 'Май 2023', 'Июнь 2023',
       'Июль 2023', 'Август 2023', 'Сентябрь 2023', 'Октябрь 2023',
       'Ноябрь 2023', 'Декабрь 2023', 'Январь 2024', 'Февраль 2024']

for column in column_names:
    data[column] = pd.to_numeric(data[column], errors='ignore')

data.info()
```

```python
# заменим все 'end' на 'стоп'
for column in data.columns[2:len(data.columns)-2]:
    data[column] = data[column].replace({'end': 'стоп'})
```

```python
# data.drop_duplicates('id')
```

```python
# функция по отсеиванию стопов
mounth_dict = {}
for i in range(len(column_names)):
    mounth_dict[column_names[i].lower()] = i
for column in data.columns[2:len(data.columns)-2]:
    temp = data[data[column] == 'стоп'].reset_index() 
    for i in range(len(temp['end_month'])):
        # print(mounth_dict[temp['end_month']])
        if mounth_dict[column.lower()] <= mounth_dict[temp['end_month'][i]]:
            data = data[data['id'] != temp['id'].to_numpy()[i]]
```
